# Filter Columns - Most Important

```python
FILTER_COLUMNS = [
    # unique identifiers
    "DUID",
    "PID",
    "DUPERSID",
    "PANEL",
    # protected attributes
    "RACEV1X",
    "RACETHX",
    "HISPANX",
    "SEX",
    # target variables
    "OBTOTV23",
    "OPTOTV23",
    "ERTOT23",
    "IPDIS23",
    "HHTOTD23",
    # demographic/socioeconomic
    "AGE23X",
    "EDUCYR",
    "MARRY53X",
    "POVCAT23",
    "POVLEV23",
    "TTLP23X",
    "INSCOV23",
    "EMPST53",
    "REGION53",
    "BORNUSA",
    "YRSINUS",
    # health
    "RTHLTH53",
    "MNHLTH53",
    "ANYLMI23",
    "PHYEXE53",
    "OFTSMK53",
    "HIBPDX",
    "DIABDX_M18",
    "ARTHDX",
    "JTPAIN31_M18",
    "CANCERDX",
    "ASTHDX",
    "CHDDX",
    "OHRTDX",
    "STRKDX",
    "CHOLDX",
    "EMPHDX",
    "CHBRON31",
    "RXTOT23",
    "PERWT23F",
    # access to care
    "HAVEUS42",
    "TYPEPE42",
    "AFRDCA42",
    "AFRDPM42",
    # costs/expenses
    "TOTEXP23",
    "TOTSLF23",
    # stats
    "VARSTR",
    "VARPSU"
]

print(f"h251 columns len: {len(data_h251.columns)}")
print(f"h251 filtered columns len: {len(FILTER_COLUMNS)}")
print(f"number of h251 discarded columns: {len(data_h251.columns) - len(FILTER_COLUMNS)}")

print("\n" + "===="* 10 + "\n")

h251_filtered = data_h251[FILTER_COLUMNS]
display(h251_filtered.head())

print("\n" + "===="* 10 + "\n")

print(f"\nh251 filtered shape: {h251_filtered.shape}")
```

```python
NULL_VALUES = [-1, -7, -8, -9, -15]

h251 = h251_filtered.replace(NULL_VALUES, np.nan).copy()

check_null_values = (h251.select_dtypes("number") < 0).sum().sum()
print(f"null values present?: {check_null_values > 0}")
print(f"null values count: {check_null_values}")
print(f"null values count after replacement: {h251.isnull().sum().sum():,}")

# print("\n" + "===="* 10 + "\n")

# """after checking output is ok to keep them"""
# num_cols = h251.select_dtypes("number").columns
# print("rest negative per column:")
# for col in num_cols:
#     negs = h251[col][h251[col] < 0].unique()
#     if len(negs) > 0:
#         print(f"  {col:<25} {sorted(negs)}")
```

# Clean Columns

```python
na_values_columns = h251.isnull().mean().sort_values(ascending=False)

print("mean of null values per column:")
for col, mean in na_values_columns.items():
    flag = ""
    if mean > 0.5:
        flag = " --> high null values"
    elif mean > 0.2:
        flag = " --> moderate null values"
    
    if mean > 0:
        print(f"  {col:<25} {mean:.2%}{flag}")
```

```python
CHRONIC_DISEASES = [
    "HIBPDX",
    "DIABDX_M18",
    "ARTHDX",
    "JTPAIN31_M18",
    "CANCERDX",
    "ASTHDX",
    "CHDDX",
    "OHRTDX",
    "STRKDX",
    "CHOLDX",
    "EMPHDX",
    "CHBRON31"
]

"""recode values 1 to 1 (yes) and 2 to 0 (no) for chronic diseases"""
for col in CHRONIC_DISEASES:
    if col in h251.columns:
        h251[col] = h251[col].map({1: 1, 2: 0})

for col in CHRONIC_DISEASES:
    if col in h251.columns:
        unique_values = h251[col].dropna().unique()
        mean_val = h251[col].mean()
        print(f"{col:<15} unique values: {unique_values} | mean: {mean_val:.1%}")
```

# Target Variable

```python
TARGETS = ["OBTOTV23", "OPTOTV23", "ERTOT23", "IPDIS23", "HHTOTD23"]

h251["TOTAL_VISITS"] = h251[TARGETS].fillna(0).clip(lower=0).sum(axis=1)
h251["HIGH_UTIL"] = (h251["TOTAL_VISITS"] >= 10).astype(int)

print(f"total persons: {len(h251)}")
print(f"high utilisation >= 10 visits: {h251['HIGH_UTIL'].sum()} ({h251['HIGH_UTIL'].mean():.1%})")
print(f"low utilisation < 10 visits: {(h251['HIGH_UTIL']==0).sum()} ({(h251['HIGH_UTIL']==0).mean():.1%})")

print("\n" + "===="* 10 + "\n")

print(f"stats total visits:\n\n {h251['TOTAL_VISITS'].describe()}")
    
```
